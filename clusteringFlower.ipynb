```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sb

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/GusttavoLopes/machine-learning/main/iris.csv'
df = pd.read_csv(url, delimiter = ',')
```

```python
sb.pairplot(df)
```

```python
X = np.array(df.drop('target',axis = 1))
```

```python
X
```

### K-means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

#### COMPARAÇÃO DAS CLASSES CRIADAS COM AS CLASSES EXISTENTES DE IRIS

```python
df['K-classes'] = kmeans.labels_
```

```python
df
```

```python
sb.pairplot(df,hue='target')
```

```python
sb.pairplot(df,'K-classes')
```
